Nesse trabalho foi feita uma análise exploratória dos dados de vendas da empresa fictícia AdventureWorks (Microsoft) como uma forma de introdução ao uso da bilbioteca Pandas.

In this work, an exploratory analysis of sales data from the fictitious company AdventureWorks (Microsoft) was carried out as a way of introducing the use of the Pandas library.

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df["Valor Venda"].sum()
```

```python
df.sample(5)
```

```python
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.sample(5)
```

```python
df["custo"].sum()
```

```python
df["lucro"] = df["Valor Venda"] - df["custo"]
```

```python
df.sample(5)
```

```python
round(df["lucro"].sum(), 2)
```

```python
#Coluna com os dias para enviar o produto a partir da venda
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.sample(5) 
```

```python
df["Tempo_envio"].dtype
```

```python
round(df.groupby("Marca")["Tempo_envio"].mean(), 2)
```

```python
#verificar valores ausentes
df.isnull().sum()
```

Lucro por ano e por data

```python
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
lucro_ano
```

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Pridutos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum()
```

```python
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
df_2009.groupby("Marca")["lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal");
```

```python
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal");
```
